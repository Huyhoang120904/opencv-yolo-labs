```python
import cv2

# Read an image
image = cv2.imread("./1028.jpg")
```

```python
# Show an image
cv2.imshow('Coffee leaf image: ' , image)
cv2.waitKey(0)
```

```python
#Write an image
cv2.imwrite('testWrite.jpg' , image)
```

```python
#Image processing


resizing = cv2.resize(image, (1080, 1080))
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurring = cv2.GaussianBlur(image, (5,5), 0)
edges = cv2.Canny(resizing, 100, 150)

# cv2.imshow('Resizing image: ' , resizing)
# cv2.imshow('Grey image: ' , grey)
# cv2.imshow('Blurring leaf image: ' , blurring)
cv2.imshow('Edges leaf image: ' , edges)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import numpy as np

canvas = np.zeros((512, 512, 3), dtype=np.uint8)

cv2.line(canvas, (0,0), (511,511), (0,255,0), 15)

cv2.imshow("Canvas with shapes", canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
